```python
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage.filters import threshold_otsu
import numpy as np

import pandas as pd
import missingno as msno

def missingno_figures(img_file):
    """
    Takes an image as input and draws it on a missingno's matrix.
    """
    img = plt.imread(img_file)
    
    img_gray = rgb2gray(img) # The threshold does not work on colored images
    threshold = threshold_otsu(img_gray)
    img_binary = (img_gray > threshold)
    
    df = pd.DataFrame(img_binary).replace(True, np.nan) # missingno requires pandas.
    msno.matrix(df)
```

- - -
### Trying the function on a simple, pixelated image

```python
car = plt.imread("carfig.jpg")
plt.axis("off")
plt.imshow(car)
```

```python
missingno_figures("carfig.jpg")
```

### Tring the function on a complex image
